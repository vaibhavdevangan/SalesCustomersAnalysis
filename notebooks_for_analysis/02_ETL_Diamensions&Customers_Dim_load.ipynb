### Select customer-related columns
### Loading dim_customer to Database 

```python
dim_customer = df[
    ["Customer ID", "Segment", "City", "State", "Country"]
].drop_duplicates()
dim_customer.shape
```

```python
dim_customer.columns
```

```python
dim_customer
```

### Renaming the Columns

```python
dim_customer.columns = [
    "customer_id",
    "segment",
    "city",
    "state",
    "country"
]
```

```python
dim_customer.shape
```

```python
dim_customer["customer_id"].isna().sum()
```

```python
dim_customer["customer_id"].nunique()
```

```python
dim_customer_clean = (
    dim_customer
    .sort_values("customer_id")
    .drop_duplicates(subset="customer_id", keep="first")
)
dim_customer_clean.shape
```

```python
dim_customer_clean["customer_id"].nunique()
```

### load dim_customer_clean to database

## dim_customer loaded successfully to Database 

```python
# Keep only unique customers in your new DataFrame
dim_customer_unique = dim_customer_clean.drop_duplicates(subset=['customer_id'])

# Fetch existing customer_ids from DB
existing_ids = pd.read_sql("SELECT customer_id FROM dim_customer", engine)

# Filter out already existing customers
dim_customer_to_insert = dim_customer_unique[
    ~dim_customer_unique['customer_id'].isin(existing_ids['customer_id'])
]

# Insert only new rows
dim_customer_to_insert.to_sql(
    'dim_customer',
    engine,
    if_exists='append',
    index=False
)
```
